```python
mean=df["Rating"].mean()
df["Rating"]=df["Rating"].fillna(mean)
```

```python
df["Android Ver"]=df["Android Ver"].fillna("4.1 and up")
```

```python
df.drop_duplicates()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df[df["Reviews"]=="3.0M"]
```

```python
df["Reviews"]=df["Reviews"].replace("3.0M",3.0)
```

```python
df["Reviews"]=df["Reviews"].astype("float")
```

```python
df["Content Rating"]=df["Content Rating"].astype("category")
```

```python
df.dtypes
```

```python
df["Installs1"]=df["Installs"].str.replace(",",'')
```

```python
df.head()
```

```python
df["Installs1"]=df["Installs1"].str.replace("+",'')
```

```python
df.head(1)
```

```python
df[df["Installs1"]=="Free"]
```

```python
df["Installs1"]=df["Installs1"].str.replace("Free",'0')
```

```python
df["Installs1"]=df["Installs1"].astype(int)
```

```python
df.drop([10472], axis=0, inplace=True)
```

# Finding the correlation

```python
corr = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, vmax=.9, linewidth=.01, square=True, annot=True)
plt.title('Coorelation between different columns')
plt.show()
```

> It is inferred from the above plot that Apps with more reviews have more installs

# Exploratory Data Analysis

# What type of category have more ratings

```python
cat=df.groupby("Category")["Rating"].mean().sort_values(ascending=False).head()
cat
```

```python
a=cat.plot(kind="pie",autopct="%1.2f%%")
plt.title("Categories based on ratings")
a.set_ylabel("")
plt.show()
```

> Almost all types of apps are equally present on googleplaystore.

# Top 5 Apps having 5.0 ratings

```python
t=df[df["Rating"]==5.0][["App","Rating"]].sort_values(by="Rating",ascending=False)[:5]
```

```python
t.plot(kind="bar")
plt.title("Free vs Paid")
plt.xlabel("Apps type")
plt.show()
```

# Number of Free and paid apps

```python
df["Type"].value_counts().plot(kind="bar")
plt.title("Free vs Paid")
plt.xlabel("Apps type")
plt.show()
```

# Top 10 Apps with highest reviews

```python
df.columns
```

```python
df.nlargest(30,"Reviews").groupby(["App"])["Reviews"].max().reset_index("App").plot(x="App",kind="bar")
plt.title("10 Apps with highest reviews")
plt.show()
```

> Facebook has the highest reviews followed by whatsapp and instagram

# Which type of app has got the max number of ratings

```python
df.columns
```

```python
g=df.groupby("Type")["Rating"].max().plot(kind="pie",autopct="%1.2f%%")
plt.title("type of app with more ratings")
g.set_ylabel("")
plt.show()
```

# Does rating depends on the type of app

# Least 10 Educational apps by ratings

```python
df.columns
```

```python
least10=df[df["Genres"].str.contains("Educational")][["App","Rating"]].sort_values(by="Rating",ascending=False).tail(10)
least10
```

# Does the installs depends on the price of the apps

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="Rating",y="Price",data=df[:5000])
plt.title("Price vs Rating")
plt.xticks(rotation=90)
plt.show()
```

>The price of the app increases the rating of the app decreases

# What genre of apps are least rated

```python
gen=df.groupby("Genres")[["Genres","Rating"]].mean().sort_values(by="Rating",ascending=False).tail(10)
gen
```

```python
sns.barplot(y=gen.index,x="Rating",data=gen)
plt.title("least rated genres")
plt.show()
```

> Mean of least rated apps are almost all equal


# Does genres affect installs

```python
aw=df.groupby("Genres")[["Category","Installs1"]].mean().sort_values(by="Installs1",ascending=False)[:20]
aw
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Genres",y="Installs1",data=aw)
plt.xticks(rotation=90)
plt.title("Genres and installs")
plt.show()
```

> Treding genres has more install and decreases otherwise
