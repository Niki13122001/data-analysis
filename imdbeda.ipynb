# Finding the correlation

```python
corr = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, vmax=.9, linewidth=.01, square=True, annot=True)
plt.title('Coorelation between different columns')
plt.show()
```

# Does Revenue and Metascore affect genre

```python
revenue_bin_edges = [0,13.27, 47.98, 113.75, 936.63]
revenue_bin_names = ['Poor', 'Marginal', 'Fair', 'High']
df['Revenue_bins'] = pd.cut(df['Revenue'], revenue_bin_edges, labels=revenue_bin_names)
df.head(5)
```

```python
rating_bin_edges = [2, 6.2, 6.8, 7.4, 9]
rating_bin_names = ['Poor', 'Okay', 'Good', 'Very Good']
df['Rating_bins'] = pd.cut(df['Rating'], rating_bin_edges, labels=rating_bin_names)
df.head()
```

```python
sns.catplot(x='Genre_count',y='Metascore', data=df, hue='Revenue_bins')
plt.title("Variation of revenue and Meatscore with Genre")
plt.show()
```

> Metascore and Revenue is high for movies which is having 3 genres
