# Finding the correlation

```python
corr = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, vmax=.9, linewidth=.01, square=True, annot=True)
plt.title('Coorelation between different columns')
plt.show()
```

> It is inferred from the above plot that the highest correlation is between isTakeaway & isIndoorSeating [the hotel which provides both is highest rated]

# How many restaurants are veg and how many are non-veg

# What is the average cost in top 5 restaurants

```python
df.columns
```

```python
top5=df.nlargest(5,"AverageCost")[["Name","AverageCost"]].sort_values(by="AverageCost",ascending=False).set_index("Name")
top5
```

```python
sns.barplot(x="AverageCost",y=top5.index,data=top5)
plt.title("Top 5 average cost")
plt.show()
```

> B Café - Shangri-La Hotel is the expensive restaurant with average cost of 4200

# Is the average cost of veg restaurants more compared to non- veg restaurants

```python
df.columns
```

```python
veg=df[df["isVegOnly"]==1][["Name","AverageCost"]]\
.sort_values(by="AverageCost",ascending=False).set_index("Name")[:7]
veg
```

```python
nonveg=df[df["isVegOnly"]==0][["Name","AverageCost"]]\
.sort_values(by="AverageCost",ascending=False).set_index("Name")[:5]
nonveg
```

```python
df1=pd.concat([veg,nonveg],keys=["veg","nonveg"])
df1
```

```python
sns.lineplot(y="AverageCost",x="Name",data=df1)
plt.xticks(rotation=90)
plt.title("veg and non veg averagecost")
plt.show()
```

> On an average non veg restaurants are quite expensive that veg restaurants
